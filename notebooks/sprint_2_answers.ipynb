## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set() # For aesthetic purposes
pd.set_option("display.max_columns", 101) # So we can see our whole DataFrame
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
```

## Load data

```python
# Training data
train = pd.read_csv("../data/train.csv").set_index('Id')
y_train = train['SalePrice']
X_train = train.drop(['SalePrice'], axis=1)

# Testing data
test = pd.read_csv("../data/train.csv").set_index('Id')
y_test = train['SalePrice']
X_test = train.drop(['SalePrice'], axis=1)
```

## Selecting features

```python
useful_features = ['LotArea', 'YearBuilt', 'YrSold']
```

```python
X = X_train[useful_features]
y = y_train
```

```python
lr = linear_model.LinearRegression()
lr.fit(X, y)
```

## Checking our performance

```python
XX = X_test[useful_features]
y_hat = lr.predict(XX)
```

```python
mean_squared_error(y_test, y_hat)
```

```python
sns.scatterplot(y_test, y_hat);
sns.lineplot()
```
